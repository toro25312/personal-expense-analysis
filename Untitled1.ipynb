```python

print("notebook")

expenses = {"monday": 1500,
            "tuesday": 1000,
            "wednesday": 600,
            "thursday": 800,
            "friday": 2000,
            }

print("expenses dict:", expenses)
print("type:", type(expenses))


print("days (keys):", list(expenses.keys()))
print("amounts (values):", list(expenses.values()))


print("items:", list(expenses.items()))
print("number of days recorded:", len(expenses))


total_spent = sum(expenses.values())
print("total spent this week:", total_spent)

def calculate_total(expense_dict):
  """return the sum of all expense amounts."""
  return sum(expense_dict.values())

def highest_expense_day(expense_dict):
  """return the day with the highest expense amount."""
  return max(expense_dict, key=expense_dict.get)

def lowest_expense_day(expense_dict):
  """return the day with the lowest expense amount."""
  return min(expense_dict, key=expense_dict.get)


print("total spent this week:", calculate_total(expenses))
print("day with highest expense:", highest_expense_day(expenses))
print("day with lowest expense:", lowest_expense_day(expenses))


budget_limit = 10000
total = calculate_total(expenses)

if total > budget_limit:
  print("you went over budget this week")
else:
 print("you are whithin the budget")

 num_days = len(expenses)
average = total / num_days

max_days = highest_expense_day(expenses)
min_days = lowest_expense_day(expenses)

print(f"days counted: {num_days}")
print(f"average per day: {average}")
print(f"max day: {max_days} -> {expenses[max_days]}")
print(f"min day: {min_days} -> {expenses[min_days]}")


import matplotlib.pyplot as plt

days = list(expenses.keys())
values = list(expenses.values())

plt.figure(figsize=(8, 5))
plt.bar(days, values)           # barras
plt.title("Weekly Expenses")
plt.xlabel("Day")
plt.ylabel("Amount (COP)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


message = (
    f"Total: {total} COP | Average/day: {average:.0f} COP | "
    f"Max: {max_days} ({expenses[max_days]}) | Min: {min_days} ({expenses[min_days]})"
)
print("Summary ->", message)
```
